```python
import pandas as pd
dirlist = pd.read_csv('directions.txt', header = None)
dirlist = dirlist.rename(columns = {0: 'Directions'})
dirlist
```

```python
dirlist[['Direction','Amount']] = dirlist.Directions.str.split(expand=True)
dirlist
```

```python
depth = 0
forward = 0
for index, row in dirlist.iterrows():
    if row['Direction'] == 'forward':
        forward += int(row['Amount'])
    elif row['Direction'] == 'down':
        depth += int(row['Amount'])
    else:
        depth -= int(row['Amount'])
final = depth*forward
final
```

```python
#Part 2: down X increases your aim by X units. up X decreases your aim by X units. forward X does two things: It increases your horizontal position by X units. It increases your depth by your aim multiplied by X.
```

```python
depth = 0
forward = 0
aim = 0
for index, row in dirlist.iterrows():
    if row['Direction'] == 'forward':
        forward += int(row['Amount'])
        depth += (aim*int(row['Amount']))
    elif row['Direction'] == 'down':
        aim += int(row['Amount'])
    else:
        aim -= int(row['Amount'])
final = depth*forward
final
```
