```python
import pandas as pd
```

```python
numlist = pd.read_csv('nums.txt', index_col = False, header = None)
```

```python
numlist = numlist.rename(columns = {0: 'Number'})
numlist['match'] = numlist['Number'] > numlist['Number'].shift()
numlist
```

```python
numlist['match'].describe()
```

```python
#Part 2: create a list of values that are the sum of the given value and the next two:
list_of_n = numlist['Number']
list_of_n
```

```python
sumList = []
for index, elem in enumerate(list_of_n[:-2]):
    sumList.append(list_of_n[index]+list_of_n[index+1]+list_of_n[index+2])
sumList
```

```python
sumList[1997]
```

```python
df2 = pd.DataFrame(sumList, columns=['SumNum'])
df2
```

```python
df2['Match'] = df2['SumNum'] > df2['SumNum'].shift()
df2
```
